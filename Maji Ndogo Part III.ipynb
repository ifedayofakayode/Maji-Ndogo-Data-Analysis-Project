### Compare auditor's scores to employee's scores
Return all rows where auditor's scores and employee's scores are the same

```python
%%sql
SELECT
    ar.location_id,
    v.record_id,
    ar.true_water_source_score AS auditor_score,
    wq.subjective_quality_score AS surveyor_score
FROM
    auditor_report AS ar
JOIN
    visits AS v
    ON v.location_id = ar.location_id
JOIN
    water_quality AS wq
    ON wq.record_id = v.record_id
WHERE 
    true_water_source_score = subjective_quality_score
    AND
    v.visit_count = 1
LIMIT 5;
```

```python
%%sql
SELECT
    ar.location_id,
    v.record_id,
    ar.true_water_source_score AS auditor_score,
    wq.subjective_quality_score AS surveyor_score
FROM
    auditor_report AS ar
JOIN
    visits AS v
    ON v.location_id = ar.location_id
JOIN
    water_quality AS wq
    ON wq.record_id = v.record_id
WHERE 
    true_water_source_score != subjective_quality_score
    AND
    v.visit_count = 1
LIMIT 5;
```

### Save the query as a CTE

```python
%%sql
WITH Incorrect_records AS(
    SELECT
    ar.location_id,
    v.record_id,
    e.employee_name,
    ar.true_water_source_score AS auditor_score,
    wq.subjective_quality_score AS surveyor_score
FROM
    auditor_report AS ar
JOIN
    visits AS v
    ON v.location_id = ar.location_id
JOIN
    water_quality AS wq
    ON wq.record_id = v.record_id
JOIN
    employee AS e
    ON e.assigned_employee_id = v.assigned_employee_id
WHERE 
    true_water_source_score != subjective_quality_score
    AND
    v.visit_count = 1
)
SELECT
    *
FROM
Incorrect_records
LIMIT 5;
```

### We calculate the  number of mistakes made by each employee
The employee names and the count of dicrepancies between their score and the auditor's score is queried and the results is returned in a table.

```python
%%sql
WITH Incorrect_records AS(
    SELECT
    ar.location_id,
    v.record_id,
    e.employee_name,
    ar.true_water_source_score AS auditor_score,
    wq.subjective_quality_score AS surveyor_score
FROM
    auditor_report AS ar
JOIN
    visits AS v
    ON v.location_id = ar.location_id
JOIN
    water_quality AS wq
    ON wq.record_id = v.record_id
JOIN
    employee AS e
    ON e.assigned_employee_id = v.assigned_employee_id
WHERE 
    true_water_source_score != subjective_quality_score
    AND
    v.visit_count = 1
)
SELECT DISTINCT
    employee_name,
    COUNT(employee_name) AS number_of_mistakes
FROM
Incorrect_records
GROUP BY employee_name
ORDER BY number_of_mistakes DESC;
```

### Convert Incorrect_records table into a View to make the code simpler to read.

```python
%%sql
CREATE VIEW Incorrect_records AS(
    SELECT
    ar.location_id,
    v.record_id,
    e.employee_name,
    ar.true_water_source_score AS auditor_score,
    wq.subjective_quality_score AS surveyor_score
FROM
    auditor_report AS ar
JOIN
    visits AS v
    ON v.location_id = ar.location_id
JOIN
    water_quality AS wq
    ON wq.record_id = v.record_id
JOIN
    employee AS e
    ON e.assigned_employee_id = v.assigned_employee_id
WHERE 
    true_water_source_score != subjective_quality_score
    AND
    v.visit_count = 1
);
SELECT
    *
FROM
    Incorrect_records
LIMIT 5;
```

### Create a CTE called error_count to return the records of employees who made above average mistakes in their records

```python
%%sql
WITH error_count AS( # This CTE calculates the number of mistakes each employee made
    SELECT DISTINCT
    employee_name,
    COUNT(employee_name) AS number_of_mistakes
FROM
Incorrect_records
GROUP BY employee_name
)
SELECT
    *
FROM
    error_count
WHERE
    number_of_mistakes > (SELECT AVG(number_of_mistakes)FROM error_count)
ORDER BY number_of_mistakes DESC;
```

### Finally create a table of each employees records and the statements which include the word "cash"

```python
%%sql
WITH error_count AS(
    SELECT DISTINCT
    employee_name,
    COUNT(employee_name) AS number_of_mistakes
FROM
Incorrect_records
GROUP BY employee_name
),

suspect_list AS(
    SELECT
    employee_name,
    number_of_mistakes
FROM
    error_count
WHERE
    number_of_mistakes > (SELECT AVG(number_of_mistakes)FROM error_count)
)
SELECT
    ir.employee_name,
    ir.location_id,
    ar.statements
FROM Incorrect_records AS ir
JOIN
    auditor_report AS ar
    ON ar.location_id = ir.location_id
WHERE employee_name IN (SELECT employee_name FROM suspect_list)
    AND ar.statements LIKE "%cash%"
```
